#### 5.- Escribe una funcion que calcule el porcentaje de un nucleotido de una secuencia y que examine la entrada para comprobar que se ha introducido un nucleotido de los cuatro tipos que hay (A, T, G o C). Devuelve el resultado con un decimal. Por ultimo comprueba que tu codigo funciona con las siguientes *assertions*. (1 pto)

```python
#Establezco mi secuencia para qe pueda leerla
def leer_secuencia(archivo):
    with open(archivo, 'r') as file:
        lines = file.readlines()
        secuencia = ''.join(line.strip() for line in lines [1:])
    return secuencia

#Calculo el pocentaje de un nucleótico en la secuencia
def calculo_porcentaje(nucleotido, secuencia):
    #Voy a comprobar que el nucleótido sea válido
    if nucleotido not in ['A', 'T', 'G', 'C']:
        raise ValueError("el nucleotido solo puede ser 'A', 'T', 'G' o 'C'")

    #Calculo el porcentaje 
    conteo = secuencia.count(nucleotido)
    porcentaje = (conteo / len(secuencia)) * 100
    return round(porcentaje, 1) #Devuelve el resultado con un decimal

#Establezco la secuencia de estudio 
secuencia = leer_secuencia('archivo.txt')

#Imprimo los resultados calculados para cada nucleótido
print("resultado para 'A':", calculo_porcentaje("A", secuencia))
print("resultado para 'T':", calculo_porcentaje("T", secuencia))
print("resultado para 'G':", calculo_porcentaje("G", secuencia))
print("resultado para 'C':", calculo_porcentaje("C", secuencia))


assert calculo_porcentaje("A",secuencia) == 29.2
assert calculo_porcentaje("T",secuencia) == 30.5
assert calculo_porcentaje("G",secuencia) == 22.0
assert calculo_porcentaje("C",secuencia) == 18.3

print("Todas las assertions funcionan correctamente.")
```

#### 6.- Devuelve la secuencia de ARN que surgiría a partir de la siguiente secuencia. Recuerda que la secuencia de ARN resultante tiene que ser complementaria a la de ADN y además, hay que cambiar un nucleótido. (1 pto)

```python
#Convierto la secuencia de ADN en ARN
def convertir_adn_arn(secuencia_adn):
    #Voy a generar un diccionario para asignar los pares de nucleótidos complementarios
    complementario = {'A': 'U', 'C': 'G', 'G': 'C', 'T': 'A'}
    secuencia_arn = ''.join(complementario[nucleotido] for nucleotido in secuencia_adn)
    return secuencia_arn

#Hago que lea la secuencia que nos dan de ADN y luego la convierta a ARN
secuencia_adn = leer_secuencia('archivo.txt')
secuencia_arn = convertir_adn_arn(secuencia_adn)

print("Secuencia ARN:", secuencia_arn)
```

#### 7.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
import re

#Enzimas y secuencias de restricción
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}

#Establezco la secuencia de estudio y elimino la primera línea ya que contiene información del nombre
def leer_secuencia(archivo):
    with open('archivo.txt', 'r') as file:
        lines = file.readlines()
        secuencia = ''.join(line.strip() for line in lines [1:])
    return secuencia

#Ahora establezco la función para encontrar los sitios de corte 
def analizar_digestion(secuencia, enzimas):
    sitios_corte = {}
    sin_sitio_corte = []
    fragmentos_finales = []

    for nombre, sitio in enzimas.items():
        sitio_regex = sitio.replace('*', '') #Para  que quite "*" 
        posiciones = [m.start() for m in re.finditer(sitio_regex, secuencia)]

        if posiciones:
            sitios_corte[nombre] =  posiciones #Para que guarde las posiciones de corte
            #Calculo los fragmentos resultantes
            cortes = [pos + sitio.index('*') for pos in posiciones] 
            cortes = [0] + cortes + [len(secuencia)]

            #Verifico los cortes:
            print("cortes calculados:", cortes)
            
            #Creo los framentos resultantes
            fragmentos = [secuencia[cortes[i]:cortes[i+1]] for i in range(len(cortes) - 1)] 
          
            #Verifico los fragmentos
            print("fragmentos resultantes:", fragmentos)
            
            fragmentos_finales.extend(fragmentos)

            #Verificación final:
            print("fragmentos finales:", fragmentos_finales)
     
        else:
            sin_sitio_corte.append(nombre) #Para añadir la enzima a las que no tienen sitio de corte

    #Ahora calculo el tamaño + número de cada fragmento
    tamaños_fragmentos = [len(fragmento) for fragmento in fragmentos_finales]
    return sitios_corte, sin_sitio_corte, fragmentos_finales, tamaños_fragmentos

#Función que ejecute todo esto:
def ejecutar_digestion(archivo):
    secuencia = leer_secuencia(archivo)
    sitios_corte, sin_sitio_corte, fragmentos, tamaños_fragmentos = analizar_digestion(secuencia, enzR)

    #Ahora hago que imprima los resultados finales para que se muestren en una función:
    print("1. Enzimas que tienen sitios y posiciones de corte:")
    for enzima, posiciones in sitios_corte.items():
        print(f" {enzima}: posiciones de corte en {posiciones}")

    if sin_sitio_corte:
        print("\n2. Enzimas sin sitio de corte:")
        for enzima in sin_sitio_corte:
            print(f" {enzima}")

    print("\n3. Resultado de la digestión: fragmentos de la secuencia")
    for i, fragmento in enumerate(fragmentos, 1):
        print(f" fragmento {i}: {fragmento}")

    print("\n4. Número y tamaño de cada fragmento:")
    for i, tamaño in enumerate(tamaños_fragmentos, 1):
        print(f" fragmento {i}: {tamaño} bases")

#Establezco la función para que se ejecute todo el análisis:
ejecutar_digestion("archivo.txt")
```

#### 8.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (2pto)
1. La secuencia traducida de proteína
2. La longitud de la secuencia de ADN original y la longitud de la secuencia de proteínas final
3. El porcentaje de aminoácidos aromáticos (en conjunto y por separado)


```python
#PUNTO 1
#Cargo la secuencia de ADN proporcionada
def leer_secuencia(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        adn_secuencia = ''.join(line.strip() for line in lines if not line.startswith(">"))
    return adn_secuencia

#Voy a definir la función para convertir la secuencia de ADN a ARN (ejercicio 6)
def convertir_adn_arn(secuencia_adn):
    #Voy a generar un diccionario para asignar los pares de nucleótidos complementarios
    complementario = {'A': 'U', 'C': 'G', 'G': 'C', 'T': 'A'}
    secuencia_arn = ''.join(complementario.get(nucleotido, "") for nucleotido in secuencia_adn)
    return secuencia_arn

#Empleo un diccionario para poder traducir los codones a sus aminoácidos correspondientes
codon_to_aminoacid = {
    'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L',
    'UCU': 'S', 'UCC': 'S', 'UCA': 'S', 'UCG': 'S',
    'UAU': 'Y', 'UAC': 'Y', 'UAA': '*', 'UAG': '*',
    'UGU': 'C', 'UGC': 'C', 'UGA': '*', 'UGG': 'W',
    'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L',
    'CCU': 'P', 'CCC': 'P', 'CCA': 'P', 'CCG': 'P',
    'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
    'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R',
    'AUU': 'I', 'AUC': 'I', 'AUA': 'I', 'AUG': 'M',
    'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T',
    'AAU': 'N', 'AAC': 'N', 'AAA': 'K', 'AAG': 'K',
    'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R',
    'GUU': 'V', 'GUC': 'V', 'GUA': 'V', 'GUG': 'V',
    'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
    'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E',
    'GGU': 'G', 'GGC': 'G', 'GGA': 'G', 'GGG': 'G'
}

#Traducto el ARN a proteína mediante empleando una función
def traducir_arn_proteina_completa(secuencia_arn):
    proteina = ""
    for i in range(0, len(secuencia_arn) - 2, 3): #Para poder recorrer la secuencia en codones o tripletes
        codon = secuencia_arn[i:i+3]
        aminoacido = codon_to_aminoacid.get(codon, "") 
        if aminoacido != "*": #Esto haría que ignore el codón de parada
            proteina += aminoacido
    return proteina

filename = 'archivo.txt'
adn_secuencia = leer_secuencia(filename)
secuencia_arn = convertir_adn_arn(adn_secuencia)
secuencia_proteina = traducir_arn_proteina_completa(secuencia_arn)

#Imprimo los resultados de la proteína traducida
print("Secuencia proteica traducida:")
print(secuencia_proteina)

#PUNTO 2
#Calculo la longitud de la secuencia de ADN orginal y la secuencia de la proteína traducida final
longitud_adn = len(adn_secuencia)
longitud_proteina = len(secuencia_proteina)

#Imprimo los resultados 
print("Longitud de la secuencia de ADN:", longitud_adn)
print("Longitud de la secuencia de proteina:", longitud_proteina) 

#PUNTO 3
#Establezco una nueva función que calcule el porcentaje de aminoácidos aromáticos, conociendo que estos son: Fenilalanina (F), Triptófano (W) y Tirosina (Y). 
def calcular_porcentaje_aminoacidos_aromaticos(secuencia_proteina):
    fenilalanina = secuencia_proteina.count('F')
    triptofano = secuencia_proteina.count('W')
    tirosina = secuencia_proteina.count('Y')

#Calculo el porcentaje de cada uno de ellos y luego de manera conjunta
    total_aminoacidos_aromaticos = fenilalanina + triptofano + tirosina
    total_aminoacidos = len(secuencia_proteina)

    if total_aminoacidos > 0:
        porcentaje_fenilalanina = (fenilalanina / total_aminoacidos) * 100 
        porcentaje_triptofano = (triptofano / total_aminoacidos) * 100 
        porcentaje_tirosina = (tirosina / total_aminoacidos) * 100 
        porcentaje_aromaticos_total = (total_aminoacidos_aromaticos / total_aminoacidos) * 100 

    else: #Por si no hay aminoácidos de algún tipo, que el porcentaje sea igual a 0.
        porcentaje_fenilalanina = 0
        porcentaje_triptofano = 0
        porcentaje_tirosina = 0
        porcentaje_aromaticos_total = 0

    return {
        "fenilalanina": porcentaje_fenilalanina,
        "triptofano": porcentaje_triptofano,
        "tirosina": porcentaje_tirosina,
        "total_aromaticos": porcentaje_aromaticos_total
    }

porcentajes_aromaticos = calcular_porcentaje_aminoacidos_aromaticos(secuencia_proteina)

#Imprimo los resultados: 
print("\nPorcentaje de aminoacidos aromaticos:")
print(f"- Fenilalanina: {porcentajes_aromaticos['fenilalanina']:.2f}%")
print(f"- Triptofano: {porcentajes_aromaticos['triptofano']:.2f}%")
print(f"- Tirosina: {porcentajes_aromaticos['tirosina']:.2f}%")
print(f"- Porcentaje TOTAL de aminoacidos aromaticos: {porcentajes_aromaticos['total_aromaticos']:.2f}%")
```
